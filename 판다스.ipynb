```python
#데이터프레임 연산


colleges = pd.read_csv('data/college.csv', index_col='INSTNM')
college_ugds = colleges.filter(like='UGDS_')



#pandas는 뱅커 반올림함
#(숫자가 양쪽의 한 가운데 있을 경우에는 가까운 짝수 쪽으로 만들어 버림)

name='Northwest-Shoals Community College'
print(college_ugds.loc[name])

print() 
print(college_ugds.loc[name].round(2))   #뱅커반올림

print()
print((college_ugds.loc[name]+0.0001).round(2))   #뱅커반올림 전 0.0001더하면 반올림함


#산술연산자
print(college_ugds+0.000501)      #더하기 연산
print(college_ugds-0.000501)        #빼기
print(college_ugds80*0.000501)    #곱하기
print(college_ugds%0.000501)      #나누기
print((college_ugds+0.000501)//0.01)     #몫을 구하는 연산자 //

```

```python
# 스크래치에서 데이터프레임 생성

import pandas as pd
import numpy as np


#1.데이터를 담은 병렬 리스트를 생성(각각 데이터프레임의 열이 되므로 동일한 형식이어야함 )
fname=['Paul','John','Richard','George']                     
lname = ['McCartney', 'Lennon', 'Starkey', 'Harrison']
birth = [1942, 1940, 1940, 1943]

#2.리스트에서 딕셔너리를 생성하고 열 이름을 리스트와 매핑한다

people = {'first': fname, 'last': lname, 'birth': birth}

#3.딕셔너리에서 데이터프레임 생성

beatles = pd.DataFrame(people)
beatles
```

```python
#다른 인덱스 지정
pd.DataFrame(people,index=['a','b','c','d'])
```
